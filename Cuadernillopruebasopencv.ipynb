# Cuadernillo de prueba para empezar con OpenCV

```python
#importamos librer√≠as necesarias
import numpy as np
import cv2
from matplotlib import pyplot as plt

```

```python
img = cv2.imread('cutiepie.jpg',0) #carga la imagen en grises 2¬∫ argumento indica grises

#mostrar imagen en el cuadernillo
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

<h2>Operaciones b√°sicas</h2>

Acceder a pixeles espec√≠ficos

OpenCV guarda imagenes en color como BGR en lugar de RGB

```python
img2 = cv2.imread('cutiepie.jpg')
px = img2[20,20]
print('valor pixel 20,20')
print(px) #Ense√±avalores BGR

# accessing RED value
print('red 10,10')
print(img2.item(10,10,2))
```

Propiedades y tipo de imagen

```python
# image shape dice n¬∫ de columnas, filas y canales
print('shape:')
print(img.shape)
print(img2.shape)

# size devuelve p√≠xeles
print('size:')
print(img.size)
print(img2.size)

#dtype devuelve tipo de imagen
print('type:')
print(img.dtype)
print(img2.dtype)
```
