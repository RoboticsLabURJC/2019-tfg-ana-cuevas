```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
import ebma
import hbma
%matplotlib notebook
```

```python
anchorframe = cv2.imread('foremanY69.png',0)
targetframe = cv2.imread('foremanY72.png',0)

newFrame, origin, direction= ebma.EBMA(targetframe, anchorframe, 16)
plt.figure('foreman')
plt.imshow(newFrame, cmap='gray')
plt.xticks([]), plt.yticks([])  # this hides the axis ticks
for spine in plt.gca().spines.values():  #hide image border
    spine.set_visible(False)
plt.show()
```

```python
fig, ax =plt.subplots()
ax.imshow(newFrame, cmap='gray')
ax.quiver(origin[1], origin[0], direction[1], direction[0], color =[1,0,0, 1])
plt.show()
```

```python
hbmaFrame, origin, direction= hbma.HBMA(targetframe, anchorframe, 16,3)
plt.figure('foreman 2')
plt.imshow(hbmaFrame, cmap='gray')
plt.xticks([]), plt.yticks([])  # this hides the axis ticks
for spine in plt.gca().spines.values():  #hide image border
    spine.set_visible(False)
plt.show()
```
